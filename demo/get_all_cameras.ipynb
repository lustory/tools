```python
import pycurl
from io import BytesIO
import json
from ffprobe import FFProbe
```

```python
url = "http://180.213.214.75:18081/open-api/device/list"
crl = pycurl.Curl()

crl.fp = BytesIO() #设置IO，py3要求使用对bytes操作的BytesIO
crl.setopt(crl.URL, url) #设置请求的Url
crl.setopt(crl.WRITEFUNCTION, crl.fp.write) #将返回的内容定向到回调函数crl.fp.write，作用是将内容写入crl.fp
crl.perform()
content = json.loads(crl.fp.getvalue()) #加载json格式内容


for i in content['data']:
    rtsp_addr = i["videoUrl"]["rtsp"]
    try:
        print(rtsp_addr)
        metadata = FFProbe(rtsp_addr)
    except OSError:
        continue
    print(metadata)
```
